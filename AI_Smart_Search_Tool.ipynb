```python
import pandas as pd

df = pd.read_csv('/content/free_course_2.csv', encoding='ISO-8859-1')


df.drop_duplicates(inplace=True)
df.dropna(subset=['Course_Title', 'Course_Description'], inplace=True)#Drop duplicated and empty rows

# Combine Title and Description for embeddings
df['Course_Text'] = df['Course_Title'] + ". " + df['Course_Description']

# Save the cleaned data
df.to_csv('cleaned_courses.csv', index=False)#Save the clean data to file

print("Cleaned data saved as '/content/free_course_2.csv'")
```

```python
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import FAISS
import pandas as pd


df = pd.read_csv('/content/cleaned_courses.csv')# clean data loaded

embeddings = HuggingFaceEmbeddings(model_name='all-MiniLM-L6-v2')#intilizing the embedding model

# Generate embeddings and store in FAISS
texts = df['Course_Text'].tolist()
metadata = [{"Course_ID": row['Course_ID'], "Course_Title": row['Course_Title']} for _, row in df.iterrows()]

# Create FAISS vector store
vector_store = FAISS.from_texts(texts, embeddings, metadatas=metadata)

# Save FAISS index locally
vector_store.save_local("faiss_index")

print("Embeddings generated and stored successfully in FAISS.")
```

```python
import gradio as gr
from langchain.chains import RetrievalQA
from langchain.llms import OpenAI
from langchain_groq import ChatGroq

# Load FAISS Index and Embeddings
embeddings = HuggingFaceEmbeddings(model_name='all-MiniLM-L6-v2')
vector_store = FAISS.load_local("faiss_index", embeddings, allow_dangerous_deserialization=True)



llm = ChatGroq(
    model_name="llama-3.3-70b-versatile",
    temperature=1,
    groq_api_key="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"
)


qa_chain = RetrievalQA.from_chain_type(  #Creating rag chain
    llm=llm,
    chain_type="stuff",
    retriever=vector_store.as_retriever()
)

# Define Query Function
def search_courses(query):  #query
    result = qa_chain.run(query)
    return result


iface = gr.Interface(  #interface of gradio
    fn=search_courses,
    inputs=gr.Textbox(label="Enter your query"),
    outputs=gr.Textbox(label="Response"),
    title="Smart Course Search Tool",
    description="Ask questions to find the best free courses on Analytics Vidhya."
)

# Launch the app

iface.launch(debug=True, share=True)
```
